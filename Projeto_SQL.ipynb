# Projeto Final SQL para Análise de Dados - EBAC

> **DADOS:**

Os dados representam informações de clientes de um banco e contam com as seguintes colunas:

*  idade = idade do cliente

* sexo = sexo do cliente (F ou M)

* dependentes = número de dependentes do cliente

* escolaridade = nível de escolaridade do clientes

* salario_anual = faixa salarial do cliente

* tipo_cartao = tipo de cartao do cliente

* qtd_produtos = quantidade de produtos comprados nos últimos 12 meses

* iteracoes_12m = quantidade de iterações/transacoes nos ultimos 12 meses

* meses_inativo_12m = quantidade de meses que o cliente ficou inativo

* limite_credito = limite de credito do cliente

* valor_transacoes_12m = valor das transações dos ultimos 12 meses

* qtd_transacoes_12m = quantidade de transacoes dos ultimos 12 meses


A tabela foi criada no **AWS Athena** junto com o **S3 Bucket** com uma versão dos dados disponibilizados  em: https://github.com/andre-marcos-perez/ebac-course-utils/tree/main/dataset

**ANÁLISE DE DADOS**

A primeira informação a ser levantada é a quantidade comparativa entre clientes dos sexo feminino e masculino, conforme segue:

**Comparação entre Homens e Mulheres**


**Query: SELECT COUNT(*), sexo FROM credito GROUP BY sexo**

**Comparação de Tipo de Cartão e Sexo**

Query: SELECT sexo, tipo_cartao, count (*) as Total
         from credito
         GROUP BY sexo, tipo_cartao;

**Comparação entre Limite de Crédito X Salario**


Query: select avg(limite_credito) as media_limite, salario_anual from credito
where salario_anual != 'na'
group by salario_anual;

**Comparação para verificação da distribuição do limite de crédito e o  perfil dos clientes**

Query: avg(limite_credito) as media_limite, sexo, salario_anual from credito where salario_anual != 'na' group by sexo, salario_anual

**CONCLUSÕES**

* A maior parte dos clientes possuem a modalidade "blue" de cartão de crédito
* A média de limite de crédito está diretamente ligada com o salário anual dos clientes
* Os homens, nesta amostra de dados, são os que possuem maior salário e por consequência, maior limite de crédito
* Apesar de haver subclasses de pessoas com  média de limite de crédito maior que 4.500,00 e salário anual maior que 40 K, percebe-se uma distribuição de cartão de crédito na modalidade "blue", que no caso em questão é a modalidade de entrada.
* É interessante efetuar levantamento para adquação das modalidades Gold, Silver e Platinum de acordo com o limite de crédito estabelecido e salário vinculado aos perfis dos clientes.
